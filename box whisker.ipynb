```python
import matplotlib.pyplot as plt

# Define the dataset
data = [6, 8, 5, 2, 7, 11, 86, 95, 502]

# Create a box-whisker plot
fig, ax = plt.subplots()
ax.boxplot(data)

# Set the title and axis labels
ax.set_title('Box-Whisker Plot')
ax.set_xlabel('Dataset')
ax.set_ylabel('Value')

# Show the plot
plt.show()
```
